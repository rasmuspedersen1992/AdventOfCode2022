```python

class Valve():
    def __init__(self,name,rate,connect):
        self.name = name
        self.rate = rate 
        self.connect = connect 
        self.distances = dict()
    def __repr__(self):
        return self.name + ', Rate: '+str(self.rate) + ', Connections: '+', '.join(self.connect)

allValves = list()
connDict = dict()
for i in curIn:
    # print(i)
    curName = i.split('Valve ')[1].split(' has flow ')[0]
    curRate = i.split('flow rate=')[1].split(';')[0]
    curConn = i.split('to valve')[1].split(', ')
    curConn[0] = curConn[0].split(' ')[1]
    # print(curName,curRate,curConn)

    newValve = Valve(curName,int(curRate),curConn)
    allValves.append(newValve)

    connDict[curName] = curConn
allValves
```

```python
def getPaths(curPath,goal,allPaths=[]):
    prevVis = curPath.split('->')
    curName = prevVis[-1]
    curDist = len(prevVis)
    # print(curDist)
    maxDist = len(connDict)
    if curDist > maxDist:
        print('max dist')
        return allPaths 
        # return curPath,curDist

    curConn = connDict[curName]
    # print(curPath,prevVis,curName,curConn)

    if goal in curConn:
        # print('Connected, distance:',curDist)
        newPath = curPath + '->' + goal 
        allPaths.append(newPath)
        # return newPath,curDist,allPaths
        return allPaths 
    else:
        allFound = []
        for newPos in curConn:
            # print(newPos,prevVis)
            # print(newPos,curConn)
            # posPaths = []
            if newPos not in prevVis:
                # print(newPos)
                newPath = curPath + '->'+newPos 
                # print('newPath',newPath)
                # sugPath,newDist = getPaths(newPath,goal) 
                newAllPaths = getPaths(newPath,goal,allPaths) 
                # print(newAllPaths)
                # print(curOut)
                if newAllPaths != None:
                    # print(newAllPaths)
                    allFound.append(newAllPaths)
        # print('allFound:',allFound)
        if len(allFound) > 0:
            # print('allFound:',allFound)
            return allFound[0]
            
# getPaths('HH','JJ') 

def getShortestPath(name1,name2):
    allPaths = getPaths(name1,name2,[])
    curMin = len(connDict)
    indexShort = -1
    for i,p in enumerate(allPaths):
        if len(p.split('->')) < curMin:
            curMin = len(p.split('->'))
            # print(p.split('->'))
            indexShort = i
    return allPaths[indexShort]

def getPathDist(path):
    return len(path.split('->'))-1

# getPathDist(getShortestPath('HH','AA'))

# getPaths('HH','AA',[]) 
```

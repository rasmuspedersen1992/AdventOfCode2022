```python
class Elf():
    def __init__(self,x,y):
        self.x = x
        self.y = y
        self.toMove = False 
        self.proposedMove = []
    
    def __repr__(self):
        return('('+str(self.x)+','+str(self.y)+')')

    def getNei(self):
        allNei = np.zeros((8,2))
        allNei[0] = [self.x - 1 , self.y - 1]
        allNei[1] = [self.x - 1 , self.y]
        allNei[2] = [self.x - 1 , self.y + 1]
        allNei[3] = [self.x , self.y - 1]
        allNei[4] = [self.x , self.y + 1]
        allNei[5] = [self.x + 1 , self.y - 1]
        allNei[6] = [self.x + 1 , self.y]
        allNei[7] = [self.x + 1 , self.y + 1]
        return allNei
    
    def getNeiInDir(self,dir):
        curNei = list()
        if (dir=='W'):
            curNei.append([self.x - 1 , self.y - 1])
            curNei.append([self.x - 1 , self.y])
            curNei.append([self.x - 1 , self.y + 1])
        elif (dir=='E'):
            curNei.append([self.x + 1 , self.y - 1])
            curNei.append([self.x + 1 , self.y])
            curNei.append([self.x + 1 , self.y + 1])
        elif (dir=='N'):
            curNei.append([self.x - 1 , self.y - 1])
            curNei.append([self.x , self.y - 1])
            curNei.append([self.x + 1 , self.y - 1])
        elif (dir=='S'):
            curNei.append([self.x - 1 , self.y + 1])
            curNei.append([self.x , self.y + 1])
            curNei.append([self.x + 1 , self.y + 1])
        return np.array(curNei)
        
    def getToMove(self,dir):
        if (dir=='N'):
            return np.array([self.x,self.y - 1])
        elif (dir=='S'):
            return np.array([self.x,self.y + 1])
        elif (dir=='W'):
            return np.array([self.x - 1,self.y])
        elif (dir=='E'):
            return np.array([self.x + 1,self.y])
```

# Function for parsing input

```python
# 
def recuIn(curI,pos):

    # Initialize list
    toReturn = list()

    # If we haven't reached the end yet
    while pos < len(curI):

        # If current position is the start of a new part
        if curI[pos] == '[':
            # Run recursively, starting in next position
            curOut,outPos = recuIn(curI,pos+1)
            # Update the current position to be the final position returned from the recursive run
            pos = outPos 
            # And append result to the list being returned
            toReturn.append(curOut)

        # If current position is the end
        elif curI[pos] == ']':
            # Return the current contents of toReturn (could be empty), and increase position by 1 to move to after the "]"
            return toReturn,pos+1  

        # If neither start or end, we are looking at the contents of a part
        else:
            # If the next symbol is a comma, move forward (this will happen after a number has been added)
            if curI[pos] == ',':
                pos += 1 
            else:
                # Let's find the end of the current number 
                # Starting from the current position
                k = pos +1
                # Running until the end
                while k < len(curI):
                    # If we find a comma or end parenthesis
                    if ((curI[k] == ',') | (curI[k] == ']')):
                        # Get the number itself 
                        toAdd = curI[pos:k]
                        # Add to the output (parsed to int)
                        toReturn.append(int(toAdd))
                        # Increase k to stop while-loop
                        k = len(curI)
                        # Increase position to end of number
                        pos += len(toAdd)
                    # If end of number hasn't been found, increase k 
                    k += 1

    return toReturn

def parFun(curI):
    return recuIn(curI,0)[0]
            
# curI = inputsTest[10]
# # curI = '[[1,2],[34]]'
# for curI in inputsTest:
#     if curI != '':
#         print('Input:',curI)
#         testOut = recuIn(curI,0)[0]
#         print('Output:',testOut)
```

```python
flagDebug = False 
def comp(thisL,thisR):
    
    if flagDebug:
        print('Compare ',thisL,'vs',thisR)

    # If both are integers
    if ((type(thisL) == int) & (type(thisR) == int)):
        if (thisL == thisR):
            # print('--Same number, continue')
            # return comp(thisL[1:],thisR[1:])
            return None
        else:
            if (thisL < thisR):
                # print('-Correct order')
                if flagDebug:
                    print('Left side is smaller, so inputs are in the right order')
                # pL = []
                return True
            else: 
                # print('-Wrong order')
                if flagDebug:
                    print('Right side is smaller, so inputs are not in the right order')
                # pL = []
                return False
                
    # If both are lists
    if ((type(thisL) == list) & (type(thisR) == list)):
        # print('Both lists')
        curPos = 0
        curMax = np.max([len(thisL),len(thisR)])
        
        # while curPos < len(thisL):
        while curPos < curMax:
            if curPos >= len(thisL):
                if flagDebug:
                    print('Left side ran out of items, so inputs are in the right order')
                return True
            if curPos >= len(thisR):
                if flagDebug:
                    print('Right side ran out of items, so inputs are not in the right order')
                return False 
            curOut = comp(thisL[curPos],thisR[curPos]) 
            
            # If it should continue, "None" is returned. 
            if (curOut != None):
                return curOut
            curPos += 1 
            # return comp(thisL[0],thisR[0])
        # print(curPos,len(thisL),len(thisR))

    # ------------
    # Mixed types
    # ------------
    # If left is list, but right is integer:
    if ((type(thisL) == list) & (type(thisR) == int)):
        # # Generate a new list to use
        # newR = list()
        # for k in range(len(thisL)):
        #     newR.append(thisR)

        newR = [thisR]

        # newR = thisR * np.ones(len(thisL)).tolist()
        return comp(thisL,newR)

    # If left is integer, but right is list:
    if ((type(thisL) == int) & (type(thisR) == list)):
        # # Generate a new list to use
        # newL = list()
        # for k in range(len(thisR)):
        #     newL.append(thisL)
        newL = [thisL]

        return comp(newL,thisR)


# Test-cases
i = 8
# curL = inputsTest[1 + i * 3]
# curR = inputsTest[0 + i * 3]
curL = inputsTest[0 + (i - 1) * 3]
curR = inputsTest[1 + (i - 1) * 3]
# curL = '[[1,1,1],3]'
# curR = '[[1,1,1],1]'
# print('Compare',curL,'vs',curR)
pL = parFun(curL)
pR = parFun(curR)

# curOut = comp(pL,pR)
comp(pL,pR)
```
